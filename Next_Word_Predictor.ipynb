```python
faqs = """
The GIK Institute is a private educational institution, named after former bureaucrat and former President of Pakistan Ghulam Ishaq Khan.
The project was delegated to Pakistani scientist, Dr. Abdul Qadeer Khan, who was one of the founding members of the institute and was once registered as an associate professor of physics.
The genesis of the Institute can be traced back to the early 1950s when Mr. Ghulam Ishaq Khan (late) became acutely aware of Pakistan's dependence on foreign expertise and imported technology. Out of his frequent interaction with the foreign and local experts emerged the idea that a centre of excellence in engineering sciences and production technology ought to be established in the country, a centre whose standards of education are comparable to those of its counterparts in the advanced countries. 
The transformation of this idea into a practical proposition took place in December 1985 when the Benevolent Community Care and Infaq Foundation donated Rs. 50 million for setting up an institute in the Khyber Pakhtunkhwa for promotion of science and technology.
Major founding donors include Nouman Benevolent Community Care, Infaq Foundation, Chiniot Anjuman-e-Islamia, Dawood Group of Industries, and the Government of the Khyber-Pakhtunkhwa which has donated land at Topi.[citation needed]
"""
```

```python
import tensorflow as tf
from tensorflow import keras 
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
Tok=Tokenizer()
```

```python
Tok.fit_on_texts([faqs])
```

```python
Tok.word_index
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
    tokenized_sentence = Tok.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python

input_sequences 
```

```python
from keras.utils import pad_sequences
```

```python
max(Tok.index_word)
```

```python
max_length=[]
for x in input_sequences:
    max_length.append(len(x))
    
```

```python
max(max_length)
```

```python
padded_text=pad_sequences(input_sequences,maxlen=57,padding='pre')
```

```python
padded_text
```

```python
X=padded_text[:,:-1]
```

```python
y=padded_text[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from keras import Sequential
from keras.layers import Embedding,LSTM,Dense,Flatten,GRU,SimpleRNN
```

```python
y.shape
```

```python
X.shape
```

```python
Model=Sequential()
Model.add(Embedding(134,100,input_length=56))
Model.add(GRU(100,return_sequences=False))
Model.add(Dense(132,activation='softmax'))
```

```python
Model.summary()
```

```python
Model.compile(loss=keras.losses.sparse_categorical_crossentropy,metrics=['accuracy'],optimizer='adam')
```

```python
Model.fit(X,y,epochs=100)
```

# Results 

```python
import time
import numpy as np
text = "The project was delegated"

for i in range(50):
  # tokenize
  token_text = Tok.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(Model.predict(padded_token_text))

  for word,index in Tok.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
